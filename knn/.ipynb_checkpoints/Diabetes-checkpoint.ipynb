```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dfx = pd.read_csv('diabetes_data/Diabetes_XTrain.csv')
dfy = pd.read_csv('diabetes_data/Diabetes_YTrain.csv')
```

```python
dfx.head()
```

```python
X_train = dfx.values
Y_train = dfy.values

X_train = X_train[:,:]
Y_train = Y_train[:,0]

X_test = pd.read_csv('diabetes_data/Diabetes_Xtest.csv').values

print(X_train.shape, Y_train.shape, X_test.shape)
print(X_test)
```

```python
X_test.head()
```

```python
def dist(x1, x2):
    return np.sqrt(sum((x1-x2)**2))
                   
def knn(X, Y, queryPoint, k=5):
    
    vals = []
    n = X.shape[0]
    
    for i in range(n):
        d = dist(queryPoint, X[i])
        vals.append([d, Y[i]])
        
    vals.sort()
    
    #Nearest/first K points
    vals = vals[:k]
    vals = np.array(vals)
    
    new_vals = np.unique(vals[:,1], return_counts=True)
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    
    return pred
```

```python
m = X_test.shape[0]
preds = []

for i in range(m):
    sol = int(knn(X_train, Y_train, X_test[i]))
    preds.append(sol)
    
preds
```
